```python
def predict_weather(cloud_cover,	sunshine,	global_radiation,	max_temp,	min_temp,	precipitation,	pressure,	snow_depth):
 
  input = pd.DataFrame([[cloud_cover,	sunshine,	global_radiation,	max_temp,	min_temp,	precipitation,	pressure,	snow_depth]], columns=['cloud_cover',	'sunshine',	'global_radiation',	'max_temp',	'min_temp',	'precipitation',	'pressure',	'snow_depth'])
  prediction = model.predict(input)[0]

  return str(prediction)
```
